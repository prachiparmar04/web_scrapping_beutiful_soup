#### Observations:
    - As Reviews_book for each Book Section does belong to one Common Html Tag. Some Review belong to <p></p> and some belong to <p class='excerpt></p>
    - So we need to make a customize code for extracting both type of TAGS.

```python
reviews_books = []

for i in range(0,5):
    
    if(len(reviews[i].find_all('p'))==3):
        for j in reviews[i].find_all('p',class_='excerpt'):
            reviews_books.append(j.text.replace('\n',''))
    else:
         
        if(len(reviews[i].find_all('p'))==4):
            c = 0
            for j in reviews[i].find_all('p'):
                c=c+1
                if(c==4):
                     reviews_books.append(j.text.replace('\n',''))
      
```

# 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The 
    scraped data should include Product Name, Price, Image URL and Average Rating

```python
amazon = requests.get('https://www.amazon.in/s?k=mobile+phones+below+20000+rupees&crid=240QRL91TRR9N&sprefix=mobile+phones+below+20000%2Caps%2C340&ref=nb_sb_ss_ts-doa-p_1_25')
amazon
```

```python

soup = BeautifulSoup(amazon.content)

mobile_sections = soup.find_all('div',class_='a-section a-spacing-medium')

mobile_sections

```

```python
products = []
urls = []
prices = []
rating = []

for i in mobile_sections:
    
    for j in i.find('div',class_='a-section aok-relative s-image-fixed-height'):
        urls.append(j.get('src'))
      
    
    for k in i.find('span',class_='a-icon-alt'):
        l =[]
        l = k.split(' ')
        rating.append(l[0])
    for l in i.find('span',class_='a-price-whole'):
        prices.append(l.replace(',',''))
    for n in i.find('span',class_='a-size-medium a-color-base a-text-normal'):
        l = []
        l = n.split('|')
        if ('with' in l[0]):
            l1 = []
            l1 = l[0].split('with')
            products.append(l1[0])
        else:
            products.append(l[0])

mobile_dict = {
    'ProductName': products,
    'Price': prices,
    'Rating': rating,
    'Image Url': urls
}

mobiles = pd.DataFrame(mobile_dict)

mobiles
```

description.

```python
ex_days = []
for i in days:
    ex_days.append(i.text)

```

```python
desc = []
text = []
short_description = []
for i in texts:
    text.append(i.text)
    li = []
    li =i.text.replace(',','.').split('with',1)
    short_description.append(li[0])
    desc.append(li[1])

descriptions= []
for i in desc:
    descriptions.append(i.replace('near','around'))
    

```

```python
tempretures = []
for i in descriptions:
    l =[]
    l =i.replace('.','').split(' ')
    count = 1
    for i in l:
        
        if(i.isdigit()):
           
            if(count == 1):
                 tempretures.append(i)
            count=count+1
        
            

```

```python
dic_weather = {
    'Day': ex_days,
    'Short Description': short_description,
    'Tempreture (F)': tempretures,
    'Description': text
}

weather_table = pd.DataFrame(dic_weather)

weather_table

```

```python
job_title = soup2.find_all('div',class_='heading_4_5 profile') 
company_name = soup2.find_all('a',class_='link_display_like_text')  
body = soup2.find_all('div',class_='item_body') 
```

```python
jobs = []
for i in job_title:
    for j in i.find('a'):
        jobs.append(j)

companies = []
for i in company_name:
    text = i.text.replace('\n','')
    companies.append(text.replace(' ',''))


```

```python
ctc = []
dates = []
count =1
for i in body[1:]:
    
    if(count%3==1):
        text = i.text.replace('\n','')
        ctc.append(text.replace(' ',''))
   
    
    elif(count%3==2):
        
        dates.append(i.text.replace(' ',''))
    
    count = count+1

```
